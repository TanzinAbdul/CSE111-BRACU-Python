```python
class Smartphone:
    
    def __init__(self, name = ''):
        self.name = name
        self.features = {}
        
    def addFeature(self, *feat):
        if self.name == '':
            print("Feature cannot be added without name")
        else:
            ind = 0
            while ind < len(feat)-1:
                if feat[ind] in self.features:
                    self.features[feat[ind]].append(feat[ind+1])
                else:
                    self.features[feat[ind]] = [feat[ind+1]]
                ind += 2
            
    def setName(self, name):
        self.name = name
        
    def printDetail(self):
        print('Phone Name:', self.name)
        for key in self.features.keys():
            print(key, end=': ')
            ind = 0
            for feat in self.features[key]:
                if ind == len(self.features[key])-1:
                    print(feat)
                else:
                    print(feat, end = ', ')
                ind += 1


s1 = Smartphone()
print("=================================")
s1.addFeature("Display", "6.1 inch")
print("=================================")
s1.setName("Samsung Note 20")
s1.addFeature("Display", "6.1 inch")
s1.printDetail()
print("=================================")
s2 = Smartphone("Iphone 12 Pro")
s2.addFeature("Display", "6.2 inch")
s2.addFeature("Ram", "6 GB")
print("=================================")
s2.printDetail()
s2.addFeature("Display", "Amoled panel")
s2.addFeature("Ram", "DDR5")
print("=================================")
s2.printDetail()
print("=================================")
```

```python
class Student:
    
    def __init__(self,name,ID):
        self.name = name
        self.ID = ID 
    
    def Details(self):
        return "Name: "+self.name+"\n"+"ID: "+self.ID+"\n" 
    
    
class CSEStudent(Student):
    
    def __init__(self, name, ID, semester):
        super().__init__(name, ID)
        self.semester = semester
        
    def Details(self):
        return super().Details() + 'Current Semester: ' + self.semester

    def addCourseWithMarks(self, *info):
        self.info = info
        self.gpa = {}
        for ind in range(0, len(self.info), 2):
            if self.info[ind+1] >=85:
                self.gpa[self.info[ind]] = 4.0
            elif 80<= self.info[ind+1] <=84:
                self.gpa[self.info[ind]] = 3.3
            elif 70<= self.info[ind+1] <=79:
                self.gpa[self.info[ind]] = 3.0
            elif 65<= self.info[ind+1] <=69:
                self.gpa[self.info[ind]] = 2.3
            elif 57<= self.info[ind+1] <=64:
                self.gpa[self.info[ind]] = 2.0
            elif 55<= self.info[ind+1] <=56:
                self.gpa[self.info[ind]] = 1.3
            elif 50<= self.info[ind+1] <=54:
                self.gpa[self.info[ind]] = 1.0
            elif self.info[ind+1] < 50:
                self.gpa[self.info[ind]] = 0.0
                
        
    def showGPA(self):
        print(self.name + ' has taken ' + str(int(len(self.info)/2)) + ' courses.')
        count = 0
        sum_gpa = 0
        for key in self.gpa.keys():
            print(key + ': ' + str(self.gpa[key]))
            count += 1
            sum_gpa += self.gpa[key]
        print('GPA of ' + self.name + ' is ' + str(round(sum_gpa/count, 2)))


Bob = CSEStudent("Bob","20301018","Fall 2020")
Carol = CSEStudent("Carol","16301814","Fall 2020")
Anny = CSEStudent("Anny","18201234","Fall 2020")
print("#########################")
print(Bob.Details())
print("#########################")
print(Carol.Details())
print("#########################")
print(Anny.Details())
print("#########################")
Bob.addCourseWithMarks("CSE111",83.5,"CSE230",73.0,"CSE260",92.5)
Carol.addCourseWithMarks("CSE470",62.5,"CSE422",69.0,"CSE460",76.5,"CSE461" ,87.0)
Anny.addCourseWithMarks("CSE340",45.5,"CSE321",95.0,"CSE370",91.0)
print("----------------------------")
Bob.showGPA()
print("----------------------------")
Carol.showGPA()
print("----------------------------")
Anny.showGPA()
```

```python
class Transport: 
    total_traveller = 0
    
    def __init__(self, name, fare):
        self.name = name 
        self.baseFare = fare

    def __str__(self):
        s = "Name: " + self.name + ", Base fare: " + str(self.baseFare) 
        return s
        
class Bus(Transport):
    bus_traveller = 0
    
    def __init__(self, name, fare):
        super().__init__(name, fare)
        print('Bus-fare of', self.name, 'is', self.baseFare, 'Taka')
        
    def addPassengerWithBags(self, *info):
        self.pass_bag = {}
        for ind in range(0, len(info), 2):
            if 3<= info[ind+1] <= 5:
                self.pass_bag[info[ind]] = self.baseFare + 60
            elif info[ind+1] > 5:
                self.pass_bag[info[ind]] = self.baseFare + 105
            else:
                self.pass_bag[info[ind]] = self.baseFare
            Bus.bus_traveller += 1
            Transport.total_traveller += 1
        
    def __str__(self):
        out = 'Name: ' + self.name  + ', Base fare: ' + str(self.baseFare) + '\nTotal Passenger(s): ' +str(Bus.bus_traveller)
        for key in self.pass_bag.keys():
            out += '\nName: ' + key + ', Fare: ' + str(self.pass_bag[key])
        return out
        
        
class Train(Transport):
    train_traveller = 0
    
    def __init__(self, name, fare):
        super().__init__(name, fare)
        print('Bus-fare of', self.name, 'is', self.baseFare, 'Taka')
        
    def addPassengerWithBags(self, *info):
        self.pass_bag = {}
        for ind in range(0, len(info), 2):
            if 3<= info[ind+1] <= 5:
                self.pass_bag[info[ind]] = self.baseFare + 60
            elif info[ind+1] > 5:
                self.pass_bag[info[ind]] = self.baseFare + 105
            else:
                self.pass_bag[info[ind]] = self.baseFare
            Train.train_traveller += 1
            Transport.total_traveller += 1
        
    def __str__(self):
        out = 'Name: ' + self.name  + ', Base fare: ' + str(self.baseFare) + '\nTotal Passenger(s): ' +str(Train.train_traveller) + '\nPassenger Details: '
        for key in self.pass_bag.keys():
            out += '\nName: ' + key + ', Fare: ' + str(self.pass_bag[key])
        return out


t1 = Bus("Volvo", 950)
print("=================================")
t1.addPassengerWithBags("David", 6, "Mike", 1, "Carol", 3)
print("=================================")
print(t1)
print("=================================")
t2 = Train("Silk City", 850)
print("=================================")
t2.addPassengerWithBags("Bob", 2, "Simon", 4)
print("=================================")
print(t2)
print("=================================")
print("Total Passengers in Transport: ", Transport.total_traveller)
```
