```python
class Team:
    
    def __init__(self, country):
        self.country = country
        self.batsman = []
        self.bowler = []
        self.players = []
        
    def addPlayer(self, *players):
        self.players = players
        for player in players:
            if player.getrole() == 'Batsman':
                print(player.getname(), 'added in the team')
                self.batsman.append([player.getname(), player.getjersey()])
            elif player.getrole() == 'Bowler':
                print(player.getname(), 'added in the team')
                self.bowler.append([player.getname(), player.getjersey()])
            else:
                print('A player cannot be added be without a role.')          
                
    def showPlayers(self):
        
        if len(self.batsman) > 0:
            print('Batsman:')
            for batsman in self.batsman:
                print('Player Name:', batsman[0], 'Jersey No:', batsman[1])
        
        if len(self.bowler) > 0:
            print('Bowler:')
            for bowler in self.bowler:
                print('Player Name:', bowler[0], 'Jersey No:', bowler[1])
        
class Player:
    def __init__(self, *info):
        if len(info) > 2:
            self.info = [info[0], info[1], str(info[2])]
        else:
            self.info = [info[0], None, str(info[1])]
    def setRole(self, role):
        self.info[1] = role
        
    def getname(self):
        name = self.info[0]
        return name
    def getrole(self):
        return self.info[1]
    def getjersey(self):
        return self.info[2]


team = Team("Bangladesh")
p1 = Player("Mahmudullah", "Batsman",30)
print("=================================")
team.addPlayer(p1)
print("=================================")
team.showPlayers()
print("=================================")
p2 = Player("Mustafizur Rahman", "Bowler",90)
p3 = Player("Tamim Iqbal", 28)
print("=================================")
team.addPlayer(p2, p3)
print("=================================")
team.showPlayers()
print("=================================")
p3.setRole("Batsman")
print("=================================")
team.addPlayer(p3)
print("=================================")
team.showPlayers()
print("=================================")
p4 = Player("Mushfiqur Rahim", "Batsman",15)
p5 = Player("Taskin Ahmed", "Bowler",3)
print("=================================")
team.addPlayer(p4,p5)
print("=================================")
team.showPlayers()
```
