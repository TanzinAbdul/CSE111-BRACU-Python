```python
class Actor:
    def __init__(self, name, rating, industry):
        self.name = name
        self.rating = rating
        self.industry = industry

    def add_movie(self, *info):
        pass

    def __str__(self):
        s = f"Name: {self.name}\nRating: {self.rating}\nIndustry: {self.industry}"
        return s

# Write your codes here.

class TomCruise(Actor):
    
    def __init__(self, name, rating, industry, oscar):
        super().__init__(name, rating, industry)
        self.oscar = oscar
        self.movies = {}
        
    def add_movie(self, *info):
        ind = 0
        while ind < len(info)-1:
            if info[ind+1] in self.movies:
                self.movies[info[ind+1]].append(info[ind])
            else:
                self.movies[info[ind+1]] = [info[ind]]
            ind += 2
            
    def oscar_status(self):
        return self.name + ' has won ' +  str(self.oscar) + ' Oscar(s)!!!'
        
    def __str__(self):
        string = ("Actor details: \nName: " + self.name + "\nRating: " +str(self.rating) + '\nIndustry: ' + self.industry + '\nOscar Win: '+ str(self.oscar)+'\nMovies:\n')
        movies = ''
        for genre in self.movies.keys():
            movies += str(genre)+ ': ' + str(self.movies[genre])+ '\n'
        return string + str(movies)
        
        
    
class DulquerSalman(Actor):
    
    def __init__(self, name, rating, industry, award):
        super().__init__(name, rating, industry)
        self.award = award
        self.movies = {}
        
    def add_movie(self, *info):
        ind = 0
        while ind < len(info)-1:
            if info[ind] in self.movies:
                self.movies[info[ind]].append(info[ind+1])
            else:
                self.movies[info[ind]] = [info[ind+1]]
            ind += 2
            
    def award_status(self):
        return str(self.name) + ' has won ' +  str(self.award) + ' award(s)!!!'
        
    def __str__(self):
        string = ("Actor details: \nName: " + self.name + "\nRating: " +str(self.rating) + '\nIndustry: ' + self.industry + '\nAward Win: '+ str(self.award)+'\nMovies:\n')
        movies = ''
        for genre in self.movies.keys():
            movies += str(genre)+ ': ' + str(self.movies[genre])+ '\n'
        return string + str(movies)



tom_cruise = TomCruise("Tom Cruise", 9.5, "Hollywood", 0)
tom_cruise.add_movie("Knight & Day", "Action", "Top Gun", "Action", "Jerry Maguire", "Romance", "Jack Reacher", "Thriller", "The Firm", "Thriller")
print('1.------------------------------------')
print(tom_cruise.oscar_status())
print('2.------------------------------------')
print(tom_cruise)
print('3.====================================')
dulquer_salman = DulquerSalman("Dulquer Salman", 9, "Malayalam", 5)
dulquer_salman.add_movie("Thriller", "Bangalore Days", "Romance", "Ustad Hotel", "Thriller", "Charlie", "Action", "Kurup", "Action", "Vikramadithyan")
print('4.------------------------------------')
print(dulquer_salman.award_status())
print('5.------------------------------------')
print(dulquer_salman)
```
