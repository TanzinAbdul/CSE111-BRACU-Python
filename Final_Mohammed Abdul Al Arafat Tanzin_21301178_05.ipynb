```python
user = ''
temp_count = 0
result = {}

while user != 'STOP':
    temp_count += 1
    user = input('Enter name and price: ')
    if user == 'STOP':
        break
    user_str = user.split(' ')
    if user_str[0] in result:
        result[user_str[0]].append(user_str[0][:2]+'_'+user_str[1]+'_'+str(temp_count))
    else:
        result[user_str[0]] = [user_str[0][0:2]+'_'+user_str[1]+'_'+str(temp_count)]
    
    
print(result)
```

```python
class Person:
    def __init__(self, name, contact):
        self.name = name
        self.contact = contact
 
    def __str__(self):
        s = f"Name: {self.name}\nContact: {self.contact}"
        return s

#Write your code here

class Doctor(Person):
    
    def __init__(self, name, contact, speciality, hospital):
        super().__init__(name, contact)
        self.speciality = speciality
        self.hospital = hospital
        self.patients = {}
        
    def add_patients(self, *info):
        for d in info:
            data = d.split(' ')
            if data[0] in self.patients:
                self.patients[data[0]].append(d)
            else:
                self.patients[data[0]] = [d]
                
    def __str__(self):
        return super().__str__() + '\nSpeciality: ' + self.speciality + '\nHospital: ' + self.hospital + '\nPatients List:\n' + str(self.patients)
 
    def show_patient_list(self, place):
        if place in self.patients:
            print('Paitents from ' + place + ':')
            for patient in self.patients[place]:
                print(patient)
        else:
            print(self.name, 'has no paitents from', place)

            
class Engineer(Person):
    
    def __init__(self, name, contact, field):
        super().__init__(name, contact)
        self.field = field
        self.projects = {}
        

    def add_projects(self, *info):
        for d in info:
            data = d.split(' ')
            if data[len(data)-1] in self.projects:
                self.projects[data[len(data)-1]].append(d)
            else:
                self.projects[data[len(data)-1]] = [d]
                
    def __str__(self):
        return super().__str__() + '\nField: ' + self.field + '\nProject List:' + str(self.projects)


 
d = Doctor("Dr Strange", "4585632", "Neurosurgery", "DMC")
d.add_patients("Dhaka Abdur Rahim", "Dhaka John Cena", "Narayanganj John Wick", "Comilla CM Punk")
print("1.=========================================================")
print(d)
print("2.=========================================================")
d.show_patient_list("Dhaka")
print("3.=========================================================")
d.show_patient_list("Rajshahi")
print("4.=========================================================")
d.add_patients("Narayanganj Naruto Uzumaki")
print("5.=========================================================")
print(d)
print("6.=========================================================")
eng = Engineer("Howard Wolowitz", "982685", "Mechanical")
eng.add_projects("Mongol Tori Bangladesh", "Duburi Bangladesh", "Curiosity Rover USA", "Perseverence USA")
print("7.=========================================================")
print(eng)
```
