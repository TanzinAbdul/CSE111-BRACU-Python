```python
from multipledispatch import dispatch

class Batsman:
    @dispatch(int, int)
    def __init__(self, runs, balls):
        self.name = 'New Batsman'
        self.runs = runs
        self.balls = balls
    @dispatch(str, int, int)
    def __init__(self, name, runs, balls):
        self.name = name
        self.runs = runs
        self.balls = balls
    def printCareerStatistics(self):
        print('Name:', self.name)
        print('Runs Scored:', self.runs, 'Balls Faced:', self.balls)
    def battingStrikeRate(self):
        self.strike = self.runs / self.balls * 100
        return self.strike
    def setName(self, name):
        self.name = name

b1 = Batsman(6101, 7380) 
b1.printCareerStatistics() 
print("============================") 
b2 = Batsman("Liton Das", 678, 773) 
b2.printCareerStatistics() 
print("----------------------------") 
print(b2.battingStrikeRate()) 
print("============================") 
b1.setName("Shakib Al Hasan") 
b1.printCareerStatistics() 
print("----------------------------") 
print(b1.battingStrikeRate())
```

```python
class Vaccine:
    
    def __init__(self, name, country, period):
        self.name = name
        self.country = country
        self.period = period
        
    def give_info(self):
        return [self.name, self.country, self.period]
        

class Person:
    
    def __init__(self, name, age, type = 'Genreal Citizen'):
        self.name = name
        self.age = age
        self.type = type
        
    def pushVaccine(self, vaccine, second_dose = None):
        if self.age < 25 and self.type != 'Student':
            print('Sorry '+ self.name+', '+ 'Minimum age for taking vaccine is 25 years now')
        else:
            if second_dose == None:
                self.vaccine_info = Vaccine.give_info(vaccine)
                print('1st dose done for', self.name)
                self.second_dose = ('Please come after '+ str(self.vaccine_info[2]) + ' days')
                self.first_dose = 'Given'
            else:
                if self.vaccine_info[0] == (Vaccine.give_info(vaccine))[0]:
                    print('2nd dose done for', self.name)
                    self.second_dose = 'Given'
                else:
                    print('Sorry '+ self.name+ ", you can't take two different vaccines")
        
    def showDetail(self):
        print('Name:', self.name, 'Age:', self.age, 'Type:', self.type)
        print('Vaccine Name:', self.vaccine_info[0])
        print('1st dose:', self.first_dose)
        print('2nd dose:', self.second_dose)


astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
