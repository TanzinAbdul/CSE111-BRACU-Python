```python
deadline = ''
deadlines = {}
dates = []
user = ''
while user != 'STOP':
    user = input('Enter deadline: ')
    if user == 'STOP':
        break
    deadline = user.split(' ')
    if deadline[2] in deadlines:
        deadlines[deadline[2]] += ([deadline[0], deadline[1]])
    else:
        deadlines[deadline[2]] = [deadline[0], deadline[1]]
        dates.append(deadline[2])

dates = sorted(dates)
for i in range(len(dates)):
    for j in range(len(dates)-i-1):
        if int(dates[j][3:5]) > int(dates[j+1][3:5]):
            dates[j], dates[j+1] = dates[j+1], dates[j]
count = 1

for deadline in dates:
    print(deadline)
    if len(deadlines[deadline]) > 2:
        i = 0
        while i < len(deadlines[deadline]):
            print(((deadlines[deadline])[i]).upper(), '||', ((deadlines[deadline])[i+1]).upper())
            i += 2
    else:
        print(((deadlines[deadline])[0]).upper(), '||', ((deadlines[deadline])[1]).upper())
    if count < len(dates):
        print('===============')
    count += 1

#INPUTS
#cse111 mid 19/11/2021
#mat110 mid 19/11/2021
#cse111 quiz3 17/12/2021
#cse320 quiz2 27/11/2021
#cse230 quiz2 17/12/2021
#STOP
```
